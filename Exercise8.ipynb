```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v_sun = 220
r_sun = 8.5
```

```python
angle = np.flip(np.concatenate((np.array([2, 4, 6]), np.arange(10, 92, 2))))

d_vel = np.array([8.57, 10.02, 11.43, 13.14, 16.95, 14.86, 14.86, 11.43, 11.43, 11.24, 14.29, 14.10, 23.05, 27.62, 31.43, 36.19, 40.76, 46.67, 53.33, 65.33, 65.14, 66.67, 66.29, 72.95, 71.43, 78.29, 78.10, 78.29, 100.76, 106.48, 108.57, 104.96, 107.05, 108.95, 115.24, 123.81, 131.24, 141.33, 152.38, 153.14, 161.71, 192.19, 172.95, 161.9])
```

```python
def del_vel_to_vel(l, del_vel):
    return v_sun*np.sin(np.deg2rad(l)) + del_vel

def l_to_r(l):
    return r_sun*np.sin(np.deg2rad(l))
```

```python
vel = del_vel_to_vel(angle[0:len(d_vel)],d_vel)
distance = l_to_r(angle)
```

```python
plt.figure(figsize=(8,6))

plt.errorbar(distance, vel, [3]*len(vel), linewidth = 0, elinewidth=1, marker='.')
plt.hlines(220, 8.3, 8.7)

plt.ylim(0,300)

plt.show()
```
